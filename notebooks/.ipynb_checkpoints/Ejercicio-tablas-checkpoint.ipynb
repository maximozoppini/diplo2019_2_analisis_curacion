```python
from collections import Counter
nlp = spacy.load('en_core_web_sm')
#doc = nlp(u"This is a sentence $ and my name is maximo.")
#doc = nlp(libro)
doc = nlp(open("../input/libro.txt").read())
words = [token.text for token in doc if token.is_stop != True and token.is_punct != True and token.is_space != True and token.is_alpha == True and token.pos_ == 'NOUN']
#words = [token.text for token in doc if token.is_stop != True]
#words = [token.text for token in doc if token.is_stop == True]

#print(words.count())
#print(words)

print(len(words))
```

```python
frecuencias = Counter(words)
comunes = frecuencias.most_common(10)
print(comunes)
```
