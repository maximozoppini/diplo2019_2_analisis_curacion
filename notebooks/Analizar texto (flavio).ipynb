### El Sombrero de tres picos, de D. Pedro A. de Alarcón (The Project Gutenberg eBook of ).

```python
#importamos paquete para manejar hipertexto.

import requests

#el texto que nos interesa leer.
r = requests.get('https://www.gutenberg.org/files/29506/29506-h/29506-h.htm')

type(r)
```

```python
#recuperamos el HTML e imprimimos.

html = r.text

print(html)
```

## Comienza limpieza.

```python
#importamos librería para extraer datos del html.

from bs4 import BeautifulSoup

#creamos objeto desde el html.

soup = BeautifulSoup(html, "html5lib")

type(soup)
```

```python
#ahora podemos manejar distinto tipo de información, por ejemplo.

soup.title
```

```python
#idem anterior pero solo el texto.

soup.title.string
```

```python
#... hipervínculos.

soup.findAll('a')[:8]
```

```python
#... o directamente el texto (de un "plumazo").

text = soup.get_text()

print(text)
```

```python
#observamos script y "estilos", intentamos eliminarlos.

for script in soup(["script", "style"]):
    script.extract()
    
text = soup.get_text()

print(text)
```

```python
#ajustamos el texto eliminando espacios en blanco, líneas en blanco, etc.

lines = (line.strip() for line in text.splitlines())

# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print(text)
```

```python
#buscamos posición "desde y hasta" para extraer el corpus.

encabezado_hasta = text.find('———————————————————5')
print('Encabezado fin:',encabezado_hasta)

pie_desde = text.find('FIN.')
print('Pie desde:',pie_desde)
```

```python
#miramos encabezado a eliminar.

encabezado_a_eliminar = text[:encabezado_hasta]
print(encabezado_a_eliminar)
```

```python
#miramos pie a eliminar.

pie_a_eliminar = text[pie_desde:]
print(pie_a_eliminar)
```

```python
#obtenemos el corpus.

corpus = text[encabezado_hasta:pie_desde]

print(corpus)
```

## Comienza: contar palabras y ordenar por frecuencia.

```python
import spacy

nlp = spacy.load("es_core_news_sm")
```

```python
#¿cuántas sentencias hay en el texto?

doc = nlp(corpus)

sentencias = [s for s in doc.sents]

print(len(sentencias))
```

```python
#chequeamos las StopWords predefinidas.

spacy_stopwords = spacy.lang.es.stop_words.STOP_WORDS

print('Cantidad de StopWords: %d' % len(spacy_stopwords))
```

```python
#eliminamos StopWords.

tokens = [token.text for token in doc if not token.is_stop]

print('Texto original: %s' % (doc))
print()
print(tokens)
```

```python
#customizamos StopWords.

mis_stopwords = ['\n', 'y', '.', 'a', '-', '!', '...', '¡', '(', ';', ')', '?', '¿', ':', ',']

for w in mis_stopwords:
    nlp.vocab[w].is_stop = True
```

```python
#eliminamos StopWords customizadas.

tokens = [token.text for token in doc if not token.is_stop]

print('Texto original: %s' % (doc))
print()
print(tokens)
```

```python
#contamos tokens.

from collections import Counter

conteo = Counter(tokens)
mayor_frecuencia = conteo.most_common(20)

print(conteo)
```

```python
print('tokens con mayor frecuencia:',mayor_frecuencia)
```
